### Model stacking

```python
# 5 models each with different number of layers and units
UNITS1 = [64, 64]
UNITS2 = [32, 32]
UNITS3 = [64, 128, 64]
UNITS4 = [32, 64, 32]
UNITS5 = [128, 128, 256, 256]

DROPOUT1 = [0.2, 0.2]
DROPOUT2 = [0.1, 0.1]
DROPOUT3 = [0.2, 0.3, 0.2]
DROPOUT4 = [0.1, 0.2, 0.1]
DROPOUT5 = [0.2, 0.3, 0.4, 0.5]

EPOCHS = 250
KERNEL_REGULARIZER = None
PATIENCE = 15

UNITS = [UNITS1, UNITS2, UNITS3, UNITS4, UNITS5]
DROPOUTS = [DROPOUT1, DROPOUT2, DROPOUT3, DROPOUT4, DROPOUT5]
VALIDATION_SPLIT = 0.1
```

```python
from RecurrentRegressor import LSTMRegressor, GRURegressor, MLPRegressor

lstm_regressors = []
gru_regressors = []
mlp_regressors = []

for i in range(1):
    lstm_regressors.append(LSTMRegressor(units=UNITS[2], dropouts=DROPOUTS[2], kernel_regularizer=KERNEL_REGULARIZER, patience=PATIENCE, validation_split=VALIDATION_SPLIT, epochs=EPOCHS))
    gru_regressors.append(GRURegressor(units=UNITS[3], dropouts=DROPOUTS[3], kernel_regularizer=KERNEL_REGULARIZER, patience=PATIENCE, validation_split=VALIDATION_SPLIT, epochs=EPOCHS))
    mlp_regressors.append(MLPRegressor(units=UNITS[2], dropouts=DROPOUTS[2], kernel_regularizer=KERNEL_REGULARIZER, patience=PATIENCE, validation_split=VALIDATION_SPLIT, epochs=EPOCHS))
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True)

# Init base estimators
base_estimators = []

for i in range(len(lstm_regressors)):
    base_estimators.append((f"MLP_{i+1}", mlp_regressors[i]))
    base_estimators.append((f"LSTM_{i+1}", lstm_regressors[i]))
    base_estimators.append((f"GRU_{i+1}", gru_regressors[i]))
    
# Init final estimator
final_estimator = LinearRegression()

for name, estimator in base_estimators:
	estimator.fit(X_train, y_train)

# Init stacking regressor
stacking_regressor = StackingRegressor(estimators=base_estimators, final_estimator=final_estimator, cv="prefit")
```

```python
# Fit stacking regressor
stacking_regressor.fit(X_train, y_train)
```

```python
y_pred = stacking_regressor.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error, mean_absolute_error

def directional_accuracy(y_true, y_pred):
	y_test_da = np.array(y_true)
	y_pred_0 = y_pred[1:] - y_test_da[:-1]
	y_test_0 = y_true[1:] - y_test_da[:-1]
	DA = (y_pred_0<0) == (y_test_0<0)
	return np.mean(DA)*100

da = directional_accuracy(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7), dpi=450)
plt.title(f"LSTM, GRU, MLP + LR: RMSE = {rmse}, MAE = {mae}, DA = {da}")
plt.plot(y_test, label="Actual")
plt.plot(y_pred, label="Predicted")

plt.legend()
plt.show()
```
